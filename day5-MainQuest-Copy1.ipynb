# 구매금액 FE : z-score (x - 평균)/표준편차 -> standard Scaler

```python
cc_df['cc_num'].value_counts()
```

```python
amt_info = cc_df.groupby('cc_num')['amt'].agg(['mean', 'std']).reset_index()
```

```python
amt_info.to_pickle('./amt_info.pkl')
```

```python
cc_df = cc_df.merge(amt_info, on='cc_num', how='left' )
```

```python
cc_df['amt_z'] = (cc_df['amt'] - cc_df['mean'])/cc_df['std'] # z-Score 값
```

```python
cc_df
```

```python
cc_df[cc_df['is_fraud'] == 1]
```

```python
cc_df.drop(['mean', 'std'], axis = 1, inplace = True) # mean과 std는 더이상 필요 없음
```

```python
#categoru가 포함된 Z-score 구하기
cat_info = cc_df.groupby(['cc_num', 'category'])['amt'].agg(['mean', 'std']).reset_index()
```

```python
cat_info.to_pickle('./cat_info.pkl')
```

```python
cc_df = cc_df.merge(cat_info, on=['cc_num', 'category'], how='left')
```

```python
cc_df['cat_amt_z'] = (cc_df['amt'] - cc_df['mean']) / cc_df['std']
```

```python
cc_df.drop(['mean', 'std'], axis = 1, inplace = True)
```

# 시간 분석 - 날자는 무시, 사용 시간대 별로 분석

```python
cc_df.info()
```

```python
cc_df['hour'] = pd.to_datetime(cc_df['trans_date_trans_time']).dt.hour
```

```python
cc_df.head()
```

cc_num 기준으로, 각 카드가, 총 몇 건이 결제가 되었으며, 이 중에 어떤 시간대에
얼마가 결제가 되었나?  -> 아침, 점심, 저녁으로 구분

```python
def hour_func(x):
    if (x >= 7) & (x < 12):
        return 'morning'
    elif (x >= 12) & (x < 18):
        return 'afternoon'
    elif (x >= 18) & (x < 22):
        return 'night'
    else:
        return 'evening'
```

```python
cc_df['hour_cat'] = cc_df['hour'].apply(hour_func)
```

```python
cc_df['hour_cat'].value_counts()
```

```python
def hour_func(x):
    if (x >= 6) & (x < 12):   #evening에 건수가 많은데, 오전 7시를 6시로 변경
        return 'morning'
    elif (x >= 12) & (x < 18):
        return 'afternoon'
    elif (x >= 18) & (x < 23): # night 건수가 상재적으로 적어서, 23시로 변경
        return 'night'
    else:
        return 'evening'
```

```python
cc_df['hour_cat'] = cc_df['hour'].apply(hour_func)
```

```python
cc_df['hour_cat'].value_counts()
```

cc_num 별로 총 몇 건의 거래가 있고? (moring, afternoon 등에 몇 건인지?)
